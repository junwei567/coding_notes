```python
#Standard structure for graph
import math
import matplotlib.pyplot as plt

x_list = list(range(-5,6,1)) 
y_list = []

for x in x_list:    
    
    y = 1 / (1 + math.exp(-x))
    y_list.append(y) 
    
plt.plot(x_list,y_list,marker="o") 
plt.xlabel("x-values") 
plt.ylabel("y-values")

plt.title("y = x^2 -4x + 3") 
plt.show()
print("ok")
```
